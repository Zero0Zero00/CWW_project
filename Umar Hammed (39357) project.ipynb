# 1. Loading my friend data into dictionary

```python
# Create the years and durations lists
years = [2011,2012,2013,2014,2015,2016,2017,2018,2019,2020]
durations = [103, 101, 99, 100, 100, 95, 95, 96, 93,90]
```

```python
# Create a dictionary with the two lists
movie_dict = dict(zip(years,durations))
```

```python
#Print the dictionary
print (movie_dict)
```

# 2. Creating a dataframe from a dictionary

```python
# Import pandas under its usual alias
import pandas as pd

movie_dict = {
    'Year': years,
    'Duration': durations}

# Create a DataFrame from the dictionary
durations_df = pd.DataFrame(movie_dict)

# Print the DataFrame
print(durations_df)
```

# 3. A visual inspection of our data

```python
# Import matplotlib.pyplot under its usual alias and create a figure
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(10,6))

# Draw a line plot of release_years and durations
plt.plot(years,durations)
plt.xlabel('Year')  # x-axis label
plt.ylabel('Duration') #y-axis label

# Create a title
plt.title('Movie Durations Over Years')

# Show the plot
plt.show()
```

# 4. Loading the rest of the data from a csv

```python
netflix_df=pd.read_csv('netflix_data.csv')

#print first few rows
netflix_df.head()
```

# 5. Filtering for movies

```python
# Subset the DataFrame for type "Movie"
netflix_df_movies_only = netflix_df[netflix_df['type'] == 'Movie']


# Select only the columns of interest
columns_of_interest = ['title', 'country', 'genre', 'release_year', 'duration']
netflix_movies_col_subset = netflix_df_movies_only[columns_of_interest]

#print the first few rows
netflix_movies_col_subset.head()
```

# 6. Creating a scatter plot

```python
# Create a figure and increase the figure size
fig = plt.figure(figsize=(12,8))

# Create a scatter plot of duration versus year
plt.scatter(years,durations)

# Create a title
plt.xlabel('Release Year')
plt.ylabel('Duration')
plt.title('Movie Durations Over Years')


# Show the plot
plt.show()
```

# 7. Digging deeper


```python
# Subset the DataFrame for type "Movie"
netflix_df_movies_only = netflix_df[netflix_df['type'] == 'Movie']

# Filter for durations shorter than 60 minutes
short_movies = netflix_df_movies_only[netflix_df_movies_only['duration'] < 60]

# Print the first 20 rows of short_movies
short_movies.head(20)
```

# 8. Marking non-feature films

```python
colors = []

# Iterate over rows of netflix_movies_col_subset
for index, row in netflix_movies_col_subset.iterrows():
    genre = row['genre']
    if 'Children' in genre:
        colors.append('blue')
    elif 'Stand-Up' in genre:
        colors.append('green')
    elif 'Documentaries' in genre:
        colors.append('red')
    else:
        colors.append('black')

# Inspect the first 10 values in your list
print(colors[:10])
```

# 9. Plotting with colors

```python
plt.style.use('fivethirtyeight')
fig = plt.figure(figsize=(12,8))

# Create a scatter plot of duration versus release_year
plt.scatter(netflix_df_movies_only['release_year'], netflix_df_movies_only['duration'], c=colors)



# Create a title and axis labels
plt.title('Scatter Plot of Movie Durations Over Years')
plt.xlabel('Release Year')
plt.ylabel('Duration')

# Show the plot
plt.show()
```

```python
# Calculate the average duration of movies for each release year
average_durations_by_year = netflix_df_movies_only.groupby('release_year')['duration'].mean()

# Calculate the percentage change in average duration between consecutive years
percentage_change = average_durations_by_year.pct_change()

# Determine if movies are getting shorter 
are_movies_getting_shorter = percentage_change.mean() < 0

print("Are movies getting shorter?", are_movies_getting_shorter)
```
